```python
def cut_img(img, x, y):

    x_center = img.size[0] / 2
    y_center = img.size[1] / 2
    new_x1 = x_center - x//2
    new_y1 = y_center - y//2
    new_x2 = x_center + x//2
    new_y2 = y_center + y//2
    new_img = img.crop((new_x1, new_y1, new_x2, new_y2))
    return new_img


```

```python
for i in range(len(fukui_graph_list)):
    fukui_graph = fukui_graph_list[i]
    coordinate_graph = coordinate_graph_list[i]
    img1 = Image.open(file_path+'sorce_code/example_data/experiment_data/red_blue_fukui_value/'+fukui_graph)
    img2 = Image.open(file_path+'sorce_code/example_data/experiment_data/Coordinate/'+coordinate_graph)
    new_x = min(img1.size, img2.size)[0]  
    new_y = min(img1.size, img2.size)[1]

    new_img1 = cut_img(img1, new_x, new_y)
    new_img2 = cut_img(img2, new_x, new_y)
    print(new_img1.size, new_img2.size)
    final_img2 = Image.blend(new_img1, new_img2,0.09)
    final_img2.save(file_path + "sorce_code/example_data/experiment_data/coordinated_red_blue_fukui_value/coordinate_%s.png"%(coordinate_graph))
    final_img2.close()
```
