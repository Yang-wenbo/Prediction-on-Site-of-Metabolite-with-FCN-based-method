```python
total_file_list = os.listdir('.')
total_file_list.sort(key=lambda x:x[:-2])
```

```python
small_count = 0
for i in range(len(total_file_list)):
    filename = total_file_list[i]
    with open (filename,'r') as f:
        text = f.readlines()
        if len(text) > 30:
            filename_G = 'total_target_' +filename+'.png'
            filename = filename[12:]
            filename_F = filename_G
            filename_C = 'total_target_divided' + filename + '.png'
            shutil.move(filepath + 'sorce_code/example_data/experiment_data/black_GSH_value/'+ filename_G ,filepath + 'sorce_code/example_data/experiment_data/large_molecule_GSH/'+ filename_G )
            shutil.move(filepath + 'sorce_code/example_data/experiment_data/red_blue_fukui_value/'+ filename_F ,filepath + 'sorce_code/example_data/experiment_data/large_molecule_fukui/'+ filename_F )
            shutil.move(filepath + 'sorce_code/example_data/experiment_data/Coordinate/'+ filename_C ,filepath + 'sorce_code/example_data/experiment_data/large_molecule_coordinate/'+ filename_C )
        else:
            small_count += 1
            pass
```
