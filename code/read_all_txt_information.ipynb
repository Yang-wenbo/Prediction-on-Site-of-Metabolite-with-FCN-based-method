```python
def read_all_information(label_file_name,neutral_Rn_txt_file,minus_1_Rn_txt_file):
    os.chdir(filepath + "sorce_code/example_data/xenosite_result/txt")###change to file directory where you save one molecule xenosite result
    with open(label_file_name,"r") as f:#label file name 是label下对应某一个label文件
        cp=f.readlines()
    GSH_value=[]
    com_number=[]
    atom_number=[]
    tem=[]
    for line in cp:
        a=line.split()
        GSH_value.append(a[1])
        tem.append(a[0])    
    for line in tem:
        name_label=line.split('.')
        com_number.append(name_label[0])
        atom_number.append(name_label[2])
        
    os.chdir(filepath + "sorce_code/example_data/nbo/nbo_0_txt")###change to file directory where you save nbo_0 txt files
    nbo_neutral=[]
    with open(neutral_Rn_txt_file,"r") as f:#打开对应的neutral_Rn_txt_file文件，这里存有nbo_Rn_neutral
        cp=f.readlines()
    for line in cp:
        a=line.split()
        nbo_neutral.append(a[2])
    nbo_neutral_array=[np.float(n) for n in nbo_neutral]

    #os.chdir("/home/user/yang/third_gjf_for_nbo_plus_1_Rn/txt/")#这里是移动到正电荷Rn_nbo_plus_txt存储
    #nbo_plus=[]
    #with open(plus_Rn_txt_file,"r") as f:#打开对应的plus_Rn_txt_file文件，这里存有nbo_Rn_plus
        #cp=f.readlines()
    #for line in cp:
        #a=line.split()
        #nbo_plus.append(a[2])
    #nbo_plus_array=[np.float(n) for n in nbo_plus]
        
    
    os.chdir(filepath + "sorce_code/example_data/nbo/nbo_-1_txt")###change to file directory where you save nbo_-1 txt files
    nbo_minus_1=[]
    fukui_main=[]
    #fukui_access = []
    with open(minus_1_Rn_txt_file,"r") as f:
        cp=f.readlines()
    for line in cp:
        a=line.split()
        
        nbo_minus_1.append(a[2])
    nbo_minus_1_array=[np.float(n) for n in nbo_minus_1]
    
    
    for i in range(len(nbo_minus_1_array)):
        a=nbo_minus_1_array[i]
        b=nbo_neutral_array[i]
        #c=nbo_plus_array[i]
        fukui_main_value=a-b
        #fukui_access_value=b-c
        
        fukui_main.append(fukui_main_value)
        #fukui_access.append(fukui_access_value)
        
    #os.chdir("/home/user/yang/second_gjf_for_Ra_opt_result/txt/")
    #change_conformation_force=[]
    
   # with open (change_conformation_file,"r") as f:
        #cp=f.readlines()
    #for line in cp:
        #a=line.split()
        
        #change_conformation_force.append(a[2])
    #change_conformation_force_array=[np.float(n) for n in change_conformation_force]
    

    #print(len(GSH_value))
    #print(len(fukui_main))
    #print(len(com_number))
    #print(len(atom_number))
    print(label_file_name)
    with open (filepath + "sorce_code/example_data/total/total_%s" %(label_file_name),"w") as f:###change to file directory where you want to saving total files 
        for j in range(len(GSH_value)):
            f.write("%s,%s,%s,%s\n"%(com_number[j],atom_number[j],GSH_value[j],fukui_main[j]))
    return
```

```python
label_file = os.listdir(filepath + "sorce_code/example_data/xenosite_result/txt")###change to file directory where you save one molecule xenosite result
label_file.sort(key=lambda x:x[:-4])
print(len(label_file))

neutral_file = os.listdir(filepath + "sorce_code/example_data/nbo/nbo_0_txt")###change to file directory where you save nbo_0 txt files
neutral_file.sort(key=lambda x:x[:-4])
print(len(neutral_file))

minus_file = os.listdir(filepath + "sorce_code/example_data/nbo/nbo_-1_txt")###change to file directory where you save nbo_-1 txt files
minus_file.sort(key=lambda x:x[:-4])
print(len(minus_file))
```

```python
for i in range(len(label_file)):
    label_file_name = label_file[i]
    neutral_file_name = neutral_file[i]
    minus_file_name = minus_file[i]
    read_all_information(label_file_name,neutral_file_name,minus_file_name)
    print("%s"%(i))
    print(label_file_name,neutral_file_name,minus_file_name)
    print("\n")
```
