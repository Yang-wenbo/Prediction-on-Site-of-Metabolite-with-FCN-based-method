```python
def read_neutral_nbo_out_to_txt(nbo_neutral_out_file_list):
    os.chdir(filepath + "sorce_code/example_data/nbo/nbo_0_out")#change file directory to gaussian nbo_0 output file 
    

    for i  in range (len(nbo_neutral_out_file_list)):
        start = []
        end = []
        
        filename = nbo_neutral_out_file_list[i]
        portion = os.path.splitext(filename)
        file = open (filename,"r")
        for num,value in enumerate(file):
            if "Summary of Natural Population Analysis:" in value :
                start.append(num)
            if "* Total *" in value:
                end.append(num)
        
        start = np.array(start)
        end  = np.array (end)
        print(filename)
        print(len(start))
        print(len(end))
        print(" ")
        
        start_point = start[0]+6
        end_point = end[0]-1
        with open (filename,"r") as f:
            cp = f.readlines()[start_point:end_point]
        
        with open (filepath + "sorce_code/example_data/nbo/nbo_0_txt/%s.txt"%(portion[0]),"w") as f:#change to the directory where you save output txt files
                   
                   for line in cp:
                        a = line.split()
                        if a[0] == "H":
                            pass
                        else:
                            
                            f.write("%s   %s   %s\n"%(a[0],a[1],a[2]))
    return
                    
        
                   
                   
  
```
